```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from PIL import Image

model_id = "vikhyatk/moondream2"
revision = "2024-03-06"
model = AutoModelForCausalLM.from_pretrained(
model_id, trust_remote_code=True, revision=revision
)
tokenizer = AutoTokenizer.from_pretrained(model_id, revision=revision)

```

FORM

```python
from PIL import Image
image = Image.open('imgs/Vision_Form.png')
display(image)
```

```python
image = Image.open('imgs/Vision_Form.png')
enc_image = model.encode_image(image)
print(model.answer_question(enc_image, "Describe this image.", tokenizer))
```

```python
image = Image.open('imgs/Vision_Form.png')
enc_image = model.encode_image(image)
print(model.answer_question(enc_image, "What is Date Start", tokenizer))
```

```python
image = Image.open('imgs/Vision_Form.png')
enc_image = model.encode_image(image)
print(model.answer_question(enc_image, "What is Date End", tokenizer))
```

```python
image = Image.open('imgs/Vision_Form.png')
enc_image = model.encode_image(image)
print(model.answer_question(enc_image, "Find All Dates and their meaning", tokenizer))
```
