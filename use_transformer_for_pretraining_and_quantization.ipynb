```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast, Seq2SeqTrainingArguments, Seq2SeqTrainer
from datasets import load_dataset

# Load Model and Tokenizer
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")

# Load Dataset
dataset = load_dataset("opus100", "en-ps")
train_dataset = dataset["train"]

# Define Tokenization Function
def tokenize_data(example):
    tokenizer.src_lang = "ps_XX"
    src_text = example["translation"]["ps"]
    tgt_text = example["translation"]["en"]
    src_tokenized = tokenizer(src_text, return_tensors="pt", max_length=128, truncation=True, padding="max_length")
    tgt_tokenized = tokenizer(tgt_text, return_tensors="pt", max_length=128, truncation=True, padding="max_length", )
    return {"input_ids": src_tokenized["input_ids"][0], "attention_mask": src_tokenized["attention_mask"][0], "labels": tgt_tokenized["input_ids"][0]}

# Tokenize Dataset
tokenized_train_dataset = train_dataset.map(tokenize_data)

# Define Training Arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="Translation" ,
    overwrite_output_dir=True,
    num_train_epochs=1,
    per_device_train_batch_size=2,
    save_steps=500,
    save_total_limit=2,
    push_to_hub=True,
    max_steps=300
)

# Create and Train the Trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
)

# Train the Model
trainer.train()

# Save Model
# trainer.save_model("output")

trainer.push_to_hub()
```

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast, Seq2SeqTrainingArguments, Seq2SeqTrainer
from datasets import load_dataset
quantization_config = GPTQConfig(
     bits=4,
     group_size=128,
     dataset="c4",
     desc_act=False,
     disable_exllama=True
)
# Load Model and Tokenizer
model = MBartForConditionalGeneration.from_pretrained("Shivam098/opt-translation",quantization_config=quantization_config,device_map='cuda')
tokenizer = MBart50TokenizerFast.from_pretrained("Shivam098/opt-translation")

# Load Dataset
dataset = load_dataset("opus100", "en-ps")
train_dataset = dataset["train"]

# Define Tokenization Function
def tokenize_data(example):
    tokenizer.src_lang = "ps_XX"
    src_text = example["translation"]["ps"]
    tgt_text = example["translation"]["en"]
    src_tokenized = tokenizer(src_text, return_tensors="pt", max_length=128, truncation=True, padding="max_length")
    tgt_tokenized = tokenizer(tgt_text, return_tensors="pt", max_length=128, truncation=True, padding="max_length", )
    return {"input_ids": src_tokenized["input_ids"][0], "attention_mask": src_tokenized["attention_mask"][0], "labels": tgt_tokenized["input_ids"][0]}

# Tokenize Dataset
tokenized_train_dataset = train_dataset.map(tokenize_data)

# Define Training Arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="opt-Translator" ,
    overwrite_output_dir=True,
    num_train_epochs=1,
    per_device_train_batch_size=2,
    save_steps=500,
    save_total_limit=2,
    push_to_hub=True,
    max_steps=300
)

# Create and Train the Trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
)

# Train the Model
trainer.train()

# Save Model
# trainer.save_model("output")

trainer.push_to_hub()
```
