```python
import requests
from bs4 import BeautifulSoup

# url of the website
doc = "https://medium.com/@amnahhmohammed/getting-started-with-lidar-92ff984ec9f3"

# getting response object
res = requests.get(doc)

# Initialize the object with the document
soup = BeautifulSoup(res.content, "html.parser")

# Get the whole body tag
body = soup.body

# Remove all <style> tags
for style in body(["style"]):
    style.decompose()

# Remove all <script> tags
for script in body(["script"]):
    script.decompose()

# Now the body tag doesn't contain any style or script tags
print(body)
```

```python
import requests
import json
from bs4 import BeautifulSoup

def get_medium_article(url):
    # Send a GET request to the Medium article URL
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the script tag containing the article content
    for script in soup.find_all('script'):
        if script.string and script.string.strip().startswith('window.__APOLLO_STATE__'):
            # Extract the JSON data
            json_str = script.string.split('=', 1)[1].strip().rstrip(';')
            data = json.loads(json_str)

            # Extract the article content
            for key, value in data.items():
                if 'content' in value and 'subtitle' in value:
                    return {
                        'title': value.get('title'),
                        'subtitle': value.get('subtitle'),
                        'content': value.get('content', {}).get('bodyModel', {}).get('paragraphs', [])
                    }

    return None

# Example usage
url = "https://medium.com/@amnahhmohammed/getting-started-with-lidar-92ff984ec9f3"
article = get_medium_article(url)

if article:
    print(f"Title: {article['title']}")
    print(f"Subtitle: {article['subtitle']}")
    for paragraph in article['content']:
        print(paragraph.get('text', ''))
else:
    print("Failed to extract article content")
```

```python

!pip install selenium

# Import necessary libraries
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Function to set up the driver
def setup_driver():
    options = Options()
    options.add_argument('-headless')
    return webdriver.Firefox(options=options)

# Function to fetch and clean web content
def fetch_clean_content(url):
    driver = setup_driver()
    try:
        driver.get(url)

        # Wait for the body to load
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

        # Remove scripts and styles
        driver.execute_script("""
            var elements = document.getElementsByTagName('script');
            while (elements[0]) elements[0].parentNode.removeChild(elements[0]);

            elements = document.getElementsByTagName('style');
            while (elements[0]) elements[0].parentNode.removeChild(elements[0]);

            elements = document.getElementsByTagName('*');
            for (var i = 0; i < elements.length; i++) {
                elements[i].removeAttribute('class');
                elements[i].removeAttribute('style');
            }
        """)

        # Get the cleaned HTML content
        cleaned_html = driver.find_element(By.TAG_NAME, 'body').get_attribute('innerHTML')

        return cleaned_html
    finally:
        driver.quit()

# Example usage
url = "https://medium.com/@sammigachuhi/a-brief-exploration-of-lidar-processing-in-python-304ba31287c3"
clean_content = fetch_clean_content(url)

print(clean_content)
```
